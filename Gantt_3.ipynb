```python
import matplotlib.pyplot as plt 
fig, gnt = plt.subplots() 
gnt.set_ylim(0, 50) 
gnt.set_xlim(0, 160) 
gnt.set_xlabel('Tempo para as atividades desenvolvidas') 
gnt.set_ylabel('Processos') 
gnt.set_yticks([15, 25, 35, 45, 55, 65, 75, 85]) 
gnt.set_yticklabels(['1ª FASE - ESTUDO DE CASO', '2ª FASE - LEVANTAMENTO', '3ª FASE - PROJETO', '4ª FASE - APRESENTAÇÃO', '5', '6', '7', '8']) 
gnt.grid(True) 
gnt.broken_barh([(40, 50)], (30, 9), facecolors =('tab:orange')) 

gnt.broken_barh([(40, 50)], (50, 9), facecolors =('tab:green')) 

gnt.broken_barh([(40, 50)], (60, 9), facecolors =('tab:grey')) 

gnt.broken_barh([(40, 50)], (70, 9), facecolors =('tab:pink')) 

gnt.broken_barh([(80, 40)], (10, 8), facecolors =('tab:purple')) 
gnt.broken_barh([(110, 10), (150, 10)], (40, 10), 
                         facecolors ='tab:blue') 
  
gnt.broken_barh([(10, 50), (100, 20), (130, 10)], (20, 9), 
                                  facecolors =('tab:red')) 
  
plt.savefig("gantt1.png")
```
