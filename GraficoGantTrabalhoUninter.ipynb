```python
import matplotlib.pyplot as plt 
fig, gnt = plt.subplots() 
gnt.set_ylim(0, 50) 
gnt.set_xlim(0, 120) 
gnt.set_xlabel('Quantidade de dias para as atividades desenvolvidas') 
gnt.set_ylabel('Atividade Extencionista I - Moisés Ferreira RU 2585910') 
gnt.set_yticks([15, 25, 35, 45, 55]) 
gnt.set_yticklabels(['4ª FASE - APRESENTAÇÃO', '3ª FASE - PROJETO E IMPLEMENTAÇÃO', '2ª FASE - LEVANTAMENTO DA PESQUISA', '1ª FASE - ESTUDO DE CASO']) 
gnt.grid(True) 
gnt.broken_barh([(20, 40)], (30, 9), facecolors =('tab:orange')) 
gnt.broken_barh([(60, 60)], (10, 8), facecolors =('tab:purple')) 
gnt.broken_barh([(0, 20)], (40, 10), 
                         facecolors ='tab:blue')   
gnt.broken_barh([(20, 30), (100, 20), (130, 10)], (20, 9), 
                                  facecolors =('tab:red')) 
  
plt.savefig("gantt1.png")
```
